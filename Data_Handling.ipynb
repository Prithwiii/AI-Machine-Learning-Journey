### **Data Preparation** ###

```python
y= df['Outcome']
y
```

```python
X= df.drop(['Outcome','HouseType', 'Area', 'District'], axis=1)
X
```

####**Encoding Data**####

```python
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder(sparse_output= False, drop='first')
area_type_encoded= encoder.fit_transform(X[['AreaType','Gender']])
encoded_df= pd.DataFrame(area_type_encoded, columns=encoder.get_feature_names_out(['AreaType','Gender']))
X = X.drop(['AreaType','Gender'], axis=1)
X= pd.concat([X.reset_index(drop=True), encoded_df.reset_index(drop=True)],axis=1)
```

```python
X
```

####**Splitting Data**####

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=41)
```
